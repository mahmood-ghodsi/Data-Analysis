## The best and worst movies

```python
from IPython.display import HTML
```

```python
#In order to find the best and worst movies, here are the most important columns
cols = ["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]
df_best = df[cols].copy()
df_best
```

```python
df.columns
```

```python
df_best['profit_musd'] = df.revenue_musd.sub(df.budget_musd)
```

```python
df_best['return'] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
#Now we want to change the name of the columns
#df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
#                   "Average Rating", "Popularity", "Profit", "ROI"]
df_best.columns = ["Poster_path", "Title","Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
df_best.set_index('Title',inplace=True)
```

```python
df_best
```

```python
#We want to find the movies with highest rating and ROI
df_best.sort_values(by='Average Rating',ascending = False)
```

```python
df_best[df_best.Votes > 1000].sort_values(by='Average Rating',ascending=False).head(20)
```

```python
df_best[df_best.Budget >= 5].sort_values(by= 'ROI',ascending=False).head(15)
```

```python
#In order to have more trustable results, it is better if fill the nan datas for budget and votes with 0:
df_best.Budget.fillna(0,inplace=True)
df_best.Votes.fillna(0,inplace=True)
```

```python
df_best.info()
```

```python
#Best and Worst movies in term of your desired parameter (average rating, popularity,...)
def best_worst(n , by, min_bud = 0, min_votes = 0,ascending=False):
    df2 = df_best.loc[(df_best.Votes >= min_votes) & (df_best.Budget >= min_bud)].sort_values(by=by,ascending=ascending).head(n)
    return HTML(df2.to_html(escape=False))
```

```python
best_worst(20,'Average Rating',min_bud=2,min_votes=200,ascending=True)
```

```python
#Best and Worst movies in terms of revenue
best_worst(5 , 'Revenue', min_bud = 0, min_votes = 0,ascending=False)
```

```python
#Best and Worst movies in terms of revenue
best_worst(n=5 , by='Budget')
```

```python
best_worst(n=5,by='Average Rating')
```

## What are the most common words in movies titles and taglines

```python
import sys
print(sys.executable)
```

```python
from wordcloud import WordCloud
```

```python
df
```

```python
titles = df.title.dropna()
taglines = df.tagline.dropna()
overviews = df.overview.dropna()
```

```python
titles
```

```python
title_corpus = ' '.join(titles)
taglines_corpus= ' '.join(taglines)
overviews_corpus = ' '.join(overviews)
```

```python
titles_dict = dict()
for word in title_corpus.split(' '):
    titles_dict[word] = titles_dict.get(word,0) + 1
```

```python
title_wordcloud = WordCloud(background_color='white',height=2000,width=4000,max_words=200).generate(title_corpus)
```

```python
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
taglines_wordcloud = WordCloud(background_color='white',height=2000,width=4000,max_words=200).generate(taglines_corpus)
plt.figure(figsize=(16,8))
plt.imshow(taglines_wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
overviews_wordcloud = WordCloud(background_color='white',height=2000,width=4000,max_words=200).generate(overviews_corpus)
plt.figure(figsize=(16,8))
plt.imshow(taglines_wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```
